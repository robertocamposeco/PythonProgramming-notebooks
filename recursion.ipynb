# Recursividad

- Algoritmos iterativos: capturan la computación en un conjunto de variables de estado que se actualizan en cada iteración del ciclo.

```python
#multiplica a por b, esto es multiplicar a en un 
#numero de b veces.
def mult_iter(a, b): 
    result = 0
    #la variable b es la variable de iteración
    while b > 0: 
        result += a
        b -= 1
    return result
```

```python
mult_iter(2, 5)
```

- Algoritmos recursivos: capturan la computación de manera recursiva, con un caso base y un paso recursivo. 

```python
#mismo proceso de multiplicación anterior pero
#con un algoritmo recursivo.
def mult_recur(a, b):
    if b == 1:
        return a
    else:
        return a + mult_recur(a, b-1)
```

```python
mult_recur(2, 5)
```

Otro ejemplo popular es el del factorial: 

```python
def factorial(n): 
    if n == 1:
        return 1 
    else:
        return n * factorial(n-1)
```

```python
factorial(3)
```

## Iteración vs. recursión

```python
def factorial_iter(n):
    prod = 1
    for i in range(1, n+1):
        prod *= i
    return prod

def factorial_rec(n): 
    if n == 1:
        return 1 
    else:
        return n * factorial_rec(n-1)
```

```python
factorial_iter(3), factorial_rec(3)
```

```python
def mover(n, origen, destino, auxiliar): 
    if n > 0: 
        mover(n-1, origen, auxiliar, destino)
        destino.append(origen.pop())
        mover(n-1, auxiliar, destino, origen)
```

```python
n = 3
A = [*range(n, 0, -1)]
B = []
C = []
print(A, B, C)
```

```python
mover(n, A, C, B)
print(A, B, C)
```

Esta es una forma alternativa: 

```python
def towers(n, fr, to, spare):
    if n == 1:
        print(f"Move disk from {fr} to {to}")
    else:
        towers(n-1, fr, spare, to)
        towers(1, fr, to, spare)
        towers(n-1, spare, to, fr)
```

```python
towers(3, "A", "C", "B")
```

## Sucesión de Fibonacci

```python
def fib(x):
    if x == 0 or x == 1:
        return 1
    else:
        return fib(x-1) + fib(x-2)
```

```python
fib(5)
```

## Palíndromos

```python
def isPalindrome(s):

    def toChars(s):
        s = s.lower()
        ans = ''
        for c in s:
            if c in 'abcdefghijklmnopqrstuvwxyz':
                ans = ans + c
        return ans

    def isPal(s):
        if len(s) <= 1:
            return True
        else:
            return s[0] == s[-1] and isPal(s[1:-1])

    return isPal(toChars(s))
```

```python
isPalindrome("ana")
```

```python
isPalindrome("abelewasiereisaweleba")
```

## Nota sobre eficiencia

```python
def fib(n):
    if n == 1: 
        print("Me llamaron por el 1")
        return 1
    elif n == 2: 
        print("Me llamaron por el 2")
        return 2 
    else:
        return fib(n-1) + fib(n-2)

fib(5)
```

Utilizando un diccionario, u otra técnica de memoria, podemos hacer más eficiente el cómputo: 

```python
def fib_efficient(n, d): 
    if n in d: 
        return d[n]
    else:
        print(f"Me llamaron por {n}")
        next = fib_efficient(n-1, d) + fib_efficient(n-2, d)
        d[n] = next
        return next
```

```python
d = {1:1, 2:2}
fib_efficient(5, d)
```
